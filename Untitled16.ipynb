```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mlxtend.frequent_patterns import apriori, association_rules

df = pd.read_csv('Hackathon_Ideal_Data.csv')
# Data Cleaning and Preprocessing
# Assuming there are no missing values for simplicity
# Convert categorical variables to appropriate data types if needed

# Exploratory Data Analysis (EDA)
plt.figure(figsize=(10, 6))
sns.countplot(x='MONTH', data=df)
plt.title('Monthly Transaction Count')
plt.xlabel('Month')
plt.ylabel('Transaction Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(x='QTY', data=df, bins=30)
plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='MONTH', y='VALUE', data=df)
plt.title('Total Value Over Time')
plt.xlabel('Month')
plt.ylabel('Total Value')
plt.show()

# Customer Segmentation using K-means clustering
# Assuming features for clustering are available
# Example: clustering based on 'QTY' and 'VALUE'
X = df[['QTY', 'VALUE']]
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='QTY', y='VALUE', hue='Cluster', data=df, palette='Set1')
plt.title('Customer Segmentation')
plt.xlabel('Quantity')
plt.ylabel('Value')
plt.legend(title='Cluster')
plt.show()
df.to_csv('Hackathon_Ideal_Data_Clustered.csv', index=False)
```
